```python
tf.abs(t1)
```

```python
t1
```

```python
#aggreagtions
import numpy as np
```

```python
t3= tf.constant(np.random.randint(0,200,size=30), dtype=tf.float32)
```

```python
t3.numpy()
```

```python
tf.size(t3), t3.shape, t3.ndim
```

```python
tf.reduce_min(t3)
```

```python
np.min(t3)
```

```python
tf.reduce_max(t3)
```

```python
np.max(t3)
```

```python
tf.reduce_mean(t3)
```

```python
np.mean(t3)
```

```python
tf.reduce_sum(t3)
```

```python
np.sum(t3)
```

```python
tf.math.reduce_std(t3)  #standard deviation using tf
```

```python
np.std(t3)  #using numpy
```

```python
tf.math.reduce_variance(t3)   #variance using tf
```

```python
np.var(t3)  #using numpy
```

```python
tf.argmax(t3)
```

```python
tf.argmin(t3)
```

```python
#type casting
tf.math.reduce_variance(tf.cast(t3,dtype=tf.float32))
```
