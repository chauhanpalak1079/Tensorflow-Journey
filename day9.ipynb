# **tf.math in Tensorflow:**

```python
import tensorflow as tf
```

```python
t1= tf.constant([2,4,6], dtype= tf.float32)
t1.numpy()
```

```python
t2= tf.constant([8,10,12], dtype = tf.float32)
t2.numpy()
```

```python
# 1. Addition
```

```python
tf.math.add(t1,t2).numpy()
```

```python
# 2. Subtraction
```

```python
tf.math.subtract(t2,t1).numpy()
```

```python
# 3. Multiplication
```

```python
tf.math.multiply(t1,t2).numpy()
```

```python
# 4. Division
```

```python
tf.math.divide(t2,t1).numpy()
```

```python
# 5. Square Root
```

```python
tf.math.sqrt(t1).numpy()
```

```python
# 6. Logarithm
```

```python
tf.math.log(t2).numpy()
```

```python
# 7. Maximum Value
```

```python
tf.math.reduce_max(t2).numpy()
```

```python
# 8. Sum of tensor
```

```python
tf.math.reduce_sum(t1).numpy()  #2+4+6
```

```python
 # 9. Argmax (index of max value)
```

```python
tf.math.argmax(t2).numpy()
```

```python
# 10. cosine of tensor
```

```python
tf.math.cos(t2).numpy()
```

```python
# 11. Exponentiation
```

```python
tf.math.exp(t1).numpy()
```

```python
# 12. Power
```

```python
tf.math.pow(t2, 4).numpy() # we can write t1 inplace of 4 too
```

**Generating Random Tensors:**

```python
t3= tf.random.uniform(shape=[2,3], maxval=10, dtype=tf.float32)
t3.numpy()
```

```python
t4 = tf.random.uniform(shape=[4,5], maxval= 100, dtype= tf.int32)
t4.numpy()
```

**Without a Seed:** Every time you run the code, TensorFlow will generate different random numbers.

**With a Seed:** The same "random" numbers are generated every time you run the code, as long as the seed value remains the same.

```python
rt1 = tf.random.uniform(shape=[3, 2], maxval=10, dtype=tf.int32)
rt1.numpy()
```

```python
rt2 = tf.random.uniform(shape=[3, 2], maxval=10, dtype=tf.int32)
rt2.numpy()
```

```python
tf.random.set_seed(42)

rt3 = tf.random.uniform(shape=[3, 2], maxval=10, dtype=tf.int32)
rt3.numpy()
```

```python
tf.random.set_seed(42)

rt4 = tf.random.uniform(shape=[3, 2], maxval=10, dtype=tf.int32)
rt4.numpy()
```
