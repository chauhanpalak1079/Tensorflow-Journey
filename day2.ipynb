**Scalar (0D tensor):**


```python
import tensorflow as tf

# Create a scalar tensor
scalar_tensor = tf.constant(42)
print("Scalar Tensor:")
scalar_tensor
```

**Vector(1D tensor):**

```python
# Create a vector tensor
vector_tensor = tf.constant([1, 2, 3, 4, 5])
print("\nVector Tensor:")
vector_tensor
```

**Matrix(3D tensor):**

```python
# Create a matrix tensor
matrix_tensor = tf.constat([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
print("\nMatrix Tensor:")
matrix_tensor
```

```python
#Changing datatype of constant matrix
matrix= tf.constant([[12,34],[23,33],[12,34],[73,22]], dtype=tf.int16)
print("\nMatrix Tensor:")
matrix
```

```python
b_matrix= tf.constant([[12.5,34],[23,33.5],[12,34],[73,22]], dtype=tf.float16)
b_matrix
```

**Note:**
- **Scalar (0D Tensor):** A single value with no dimensions, representing a simple number (e.g., `42`).
- **Vector (1D Tensor):** A one-dimensional array, representing a list of values (e.g., `[1, 2, 3]`).
- **Matrix (2D Tensor):** A two-dimensional array with rows and columns, representing a grid of values (e.g., `[[1, 2], [3, 4]]`).
